```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings(action='ignore')
```

```python
df = pd.DataFrame(b)
df
```

**MAE**

```python
df_medias = df[['y1_x2', 'y2_x2', 'y1_x5', 'y2_x5'] + [col_media for col_media in df.columns if 'media_mae' in col_media ]]
```

```python
df_medias
```

```python
df['menor_media_shapeffects'] = np.nan
columns_not_shapeffects = [col_media for col_media in df.columns if 'media_mae' in col_media and 'shapeffects' not in col_media]
for idx, row in df_medias.iterrows():
    df_medias.loc[idx, 'menor_media_shapeffects'] = row[[col_media for col_media in df.columns if 'media_mae_shapeffects' in col_media ]].min()
    for col in columns_not_shapeffects:
        df_medias.loc[idx, ('difference_'+col).replace("media_mae_", "")] = df_medias.loc[idx, 'menor_media_shapeffects'] - df_medias.loc[idx, col]
```

```python
df_medias
```

```python
import matplotlib.colors as mcolors
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_powershap, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:blue'], label='Powershap')
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_borutashap, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:orange'], label='Borutashap')
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_shapicant, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:green'], label='Shapicant')
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_lasso_00001, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:red'], label='Lasso 0.0001')
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_boruta, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:olive'], label='Boruta')
ax.scatter(df_medias.menor_media_shapeffects, df_medias.media_mae_pimp, facecolors='none', edgecolors = mcolors.TABLEAU_COLORS['tab:cyan'], label='PIMP')
lims = [
    np.min([ax.get_xlim(), ax.get_ylim()]),  # min of both axes
    np.max([ax.get_xlim(), ax.get_ylim()]),  # max of both axes
]

# now plot both limits against eachother
ax.plot(lims, lims, 'k-', alpha=0.75, zorder=0)
ax.set_aspect('equal')
ax.set_xlim(lims)
ax.set_ylim(lims)
ax.legend()
ax.set_xlabel('Mean MAE SHAPEffects')
ax.set_ylabel('Mean MAE every other method')
```

```python
pd.options.display.max_rows = None
serie_aux = df[(df.y1_x2 == -0.1) & (df.y1_x5 == 0.1) & (df.y2_x2 == -0.04) & (df.y2_x5 == -0.25)].squeeze()
serie_aux
```

```python
dict_final = {}
for method in ['powershap', 'borutashap', 'shapicant', 'boruta', 'pimp', 'lasso_0001', 'shapeffects_025_075', 'shapeffects_02_08', 'shapeffects_015_085', 'shapeffects_01_09', 'shapeffects_005_095']:
    dict_final[method] = {}
    for medida in ['mae', 'rmse', 'r2']:
        for medida2 in ['media', 'std', 'max', 'min']:
            dict_final[method][medida+"_"+medida2] = serie_aux[medida2+"_"+medida+"_"+method]
```
